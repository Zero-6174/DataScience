### Grandient Descent Boosting - Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.metrics import r2_score, log_loss, accuracy_score
```

### Feature Encoding & Categorical Encoding

```python
categorical = []

for col, value in attrition.iteritems():
    if value.dtype == 'object':
        categorical.append(col)

numerical = attrition.columns.difference(categorical)
print("categorical: \n", categorical)
print("numerical: \n", numerical)
```

```python
attrition_cat = attrition[categorical]
attrition_cat.head()
```

```python
attrition_cat = attrition_cat.drop("Attrition", axis = 1)
```

```python
attrition_cat_dum = pd.get_dummies(attrition_cat)
```

```python
attrition_num = attrition[numerical]
```

```python
attrition_final = pd.concat([attrition_num, attrition_cat_dum], axis = 1)
attrition_final.shape
```

### Target variable

```python
target_map = {'Yes': 1, 'No' : 0}

target = attrition['Attrition']
target = target.apply(lambda x: target_map[x])
target.head(3)
```

### Split the Train & Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(attrition_final, target, train_size=0.75, random_state=0)
```

### Build the model - GBM Classifier

```python
gb = GradientBoostingClassifier()
gb.get_params()
```

```python
gb.fit(X_train, y_train)
```

```python
y_pred_test = gb.predict(X_test)
```

```python
gb_pred = gb.predict_proba(X_test)
#print(gb_pred)
```

```python
print(accuracy_score(y_test, y_pred_test))
```
